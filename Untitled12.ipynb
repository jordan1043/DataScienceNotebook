```python
import yfinance as yf
import pandas as pd

tesla_data = yf.download('TSLA', start='2010-06-29', end='2021-06-14')
tesla_data.reset_index(inplace=True)
tesla_data.head()  # Muestra las primeras 5 filas
```

```python
import requests
from bs4 import BeautifulSoup

url_tesla = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
html_tesla = requests.get(url_tesla, headers={"User-Agent": "Mozilla/5.0"}).text
soup_tesla = BeautifulSoup(html_tesla, "html.parser")

# Buscar la tabla correcta por contenido
tables = soup_tesla.find_all("table")
for table in tables:
    if "Tesla Quarterly Revenue" in str(table):
        tesla_revenue = pd.read_html(str(table))[0]
        break

# Limpiar datos
tesla_revenue.columns = ['Date', 'Revenue']
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != '$-']
tesla_revenue['Revenue'] = tesla_revenue['Revenue'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)

tesla_revenue.tail()  # Muestra las Ãºltimas 5 filas
```

```python
gme_data = yf.download('GME', start='2010-06-29', end='2021-06-14')
gme_data.reset_index(inplace=True)
gme_data.head()  # Muestra las primeras 5 filas
```
